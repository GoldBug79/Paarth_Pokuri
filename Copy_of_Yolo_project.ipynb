```python
import os, json, cv2, numpy as np
from pathlib import Path
from tqdm import tqdm

def mask_to_polygons(mask):
    # mask: HxW uint8 (0 or 255)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    polys = []
    for cnt in contours:
        if cnt.shape[0] < 3:
            continue
        # simplify (approxPolyDP) if desired:
        epsilon = 0.002 * cv2.arcLength(cnt, True)
        approx = cv2.approxPolyDP(cnt, epsilon, True)
        pts = approx.squeeze().tolist()
        if len(pts) < 3:
            continue
        # flatten to [x1,y1,x2,y2,...]
        flat = [float(xy) for p in pts for xy in p]
        polys.append(flat)
    return polys

# Build COCO skeleton
coco = {
    "info": {"description": "model predictions"},
    "licenses": [],
    "images": [],
    "annotations": [],
    "categories": []
}

# define categories (ids must match Labellerr project categories OR you will remap)
# Replace names/order with your classes
categories = [
    {"id": 1, "name": "pedestrian"},
    {"id": 2, "name": "vehicle"}
]
coco['categories'] = categories
cat_name_to_id = {c['name']: c['id'] for c in categories}

ann_id = 1
image_id = 1

# iterate over results and build entries
for res in tqdm(results):
    # results have attributes: orig_img (numpy array) or path
    # try to get image path
    img_path = None
    if hasattr(res, 'path') and res.path is not None:
        img_path = str(res.path)
    elif hasattr(res, 'orig_img') and res.orig_img is not None:
        # can't get filename; skip including image record with filename
        img_path = None

    h, w = None, None
    if hasattr(res, 'orig_shape') and res.orig_shape is not None:
        # (h,w)
        h, w = res.orig_shape[:2]
    else:
        # fallback: read image from disk if path available
        if img_path and os.path.exists(img_path):
            img = cv2.imread(img_path)
            h, w = img.shape[:2]

    # add image record
    file_name = os.path.basename(img_path) if img_path else f"img_{image_id}.jpg"
    coco['images'].append({"id": image_id, "file_name": file_name, "width": w, "height": h})

    # get masks: many ultralytics versions expose res.masks.data (N,H,W) or res.masks.masks
    masks = None
    if hasattr(res, 'masks') and res.masks is not None:
        try:
            # some versions: res.masks.data tensor (N,H,W)
            md = res.masks.data.cpu().numpy()
            masks = md  # numpy array
        except Exception:
            # other versions: res.masks.masks is already list/np
            try:
                masks = res.masks.masks  # might be list/ndarray
                masks = np.array(masks)
            except Exception:
                masks = None

    # fallback: if no masks, you can also convert predicted polygon if available (res.boxes, res.boxes.xywh) — here we skip
    if masks is None:
        image_id += 1
        continue

    # masks shape: (N, H, W) where values are 0..1 or 0/255
    for i in range(masks.shape[0]):
        mask = masks[i]
        # ensure uint8 0/255
        if mask.max() <= 1.0:
            mask_u = (mask * 255).astype('uint8')
        else:
            mask_u = mask.astype('uint8')
        # threshold to binary
        _, mask_bin = cv2.threshold(mask_u, 127, 255, cv2.THRESH_BINARY)
        polys = mask_to_polygons(mask_bin)
        if not polys:
            continue

        # assume class id available as res.boxes.cls or res.boxes.data? try to get class index
        cls_idx = None
        try:
            # many versions store boxes in res.boxes and each box has .cls
            cls_idx = int(res.boxes.data[i, 5]) if hasattr(res.boxes, 'data') else None
        except Exception:
            cls_idx = None

        # if class index not found default to 1
        coco_cat_id = 1
        if cls_idx is not None:
            # Ultralytics class index typically 0-based -> convert to your COCO category ids
            coco_cat_id = cls_idx + 1

        for seg in polys:
            # compute bbox from segmentation polygon
            xs = seg[0::2]
            ys = seg[1::2]
            x_min, y_min = min(xs), min(ys)
            x_max, y_max = max(xs), max(ys)
            bbox = [x_min, y_min, x_max - x_min, y_max - y_min]
            area = float(Polygon(np.array(seg).reshape(-1,2)).area)

            ann = {
                "id": ann_id,
                "image_id": image_id,
                "category_id": coco_cat_id,
                "segmentation": [seg],   # list of polygons
                "area": area,
                "bbox": bbox,
                "iscrowd": 0,
                "score": float(0.99)   # optional score field
            }
            coco['annotations'].append(ann)
            ann_id += 1

    image_id += 1

# save
with open(OUT_JSON, 'w') as f:
    json.dump(coco, f)
print("Saved COCO JSON to", OUT_JSON)
```

```python
from ultralytics import YOLO
import supervision as sv
import cv2

# Load trained YOLOv8 segmentation model
model = YOLO("/content/drive/MyDrive/project_data/runs/seg_exp/weights/best.pt")

# Initialize ByteTrack tracker
tracker = sv.ByteTrack()

# Annotators for visualization
box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

# Input and output videos
video_path = "/content/drive/MyDrive/project_data/test1.mp4"
out_path   = "/content/drive/MyDrive/project_data/tracked_output1.mp4"

cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("❌ Cannot open input video")
else:
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    out = cv2.VideoWriter(out_path, cv2.VideoWriter_fourcc(*"mp4v"), fps, (width, height))

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_count += 1

        # Run YOLOv8 segmentation
        results = model(frame, imgsz=640, conf=0.25)[0]

        # Convert YOLO results to Supervision Detections
        detections = sv.Detections.from_ultralytics(results)

        # Track objects
        tracked = tracker.update_with_detections(detections)

        # Generate labels for tracked objects
        labels = [
            f"#{tracker_id} {model.names[int(class_id)]} {conf:.2f}"
            for (x1, y1, x2, y2), conf, class_id, tracker_id
            in zip(tracked.xyxy, tracked.confidence, tracked.class_id, tracked.tracker_id)
        ]

        # Annotate frame
        annotated_frame = box_annotator.annotate(scene=frame.copy(), detections=tracked)
        annotated_frame = label_annotator.annotate(scene=annotated_frame, detections=tracked, labels=labels)

        # Save frame
        out.write(annotated_frame)

        if frame_count % 30 == 0:
            print(f"Processed {frame_count} frames...")

    cap.release()
    out.release()
    print(f"✅ Tracking complete! Saved at: {out_path}")
```

```python
%%writefile app.py
import streamlit as st
from ultralytics import YOLO
import supervision as sv
import cv2
import tempfile
import json

st.title("YOLOv8 + ByteTrack Video Tracking Demo")

uploaded_video = st.file_uploader("Upload a video", type=["mp4", "avi", "mov"])
if uploaded_video:
    tfile = tempfile.NamedTemporaryFile(delete=False)
    tfile.write(uploaded_video.read())

    model = YOLO("/content/drive/MyDrive/project_data/runs/seg_exp/weights/best.pt")  # Change path appropriately
    tracker = sv.ByteTrack()

    cap = cv2.VideoCapture(tfile.name)
    results_json = []
    frame_id = 0

    progress_text = st.empty()
    progress_bar = st.progress(0)

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame, imgsz=640, conf=0.25)[0]
        detections = sv.Detections.from_ultralytics(results)
        tracked = tracker.update_with_detections(detections)  # Correct method here

        for i in range(len(tracked)):
            res = {
                "frame": frame_id,
                "track_id": int(tracked.tracker_id[i]),
                "class": int(tracked.class_id[i]),
                "bbox": [float(x) for x in tracked.xyxy[i]],
                "score": float(tracked.confidence[i]),
            }
            results_json.append(res)

        frame_id += 1
        progress_text.text(f"Processing frame {frame_id} of {total_frames}")
        progress_bar.progress(min(frame_id / total_frames, 1.0))

    cap.release()

    json_path = tfile.name + "_results.json"
    with open(json_path, "w") as f:
        json.dump(results_json, f)

    st.success("Tracking complete!")
    st.download_button("Download results.json", open(json_path, "rb").read(), file_name="results.json")
```
